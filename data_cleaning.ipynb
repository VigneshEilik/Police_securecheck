```python
#checking null values

# data.isna().sum()
data['search_type'].isna().sum()
```

```python
# fill the missing values only for 'search_type' column

data['search_type'].fillna('unknown', inplace=True)
```

```python
data.info()
```

```python
import datetime
data['search_conducted'] = data['search_conducted'].astype(bool)
data['is_arrested'] = data['is_arrested'].astype(bool)
data['drugs_related_stop'] = data['drugs_related_stop'].astype(bool)
```

```python
import pandas as pd

# Change dtype of 'stop_date' to datetime64[ns] (date) and 'stop_time' to datetime64[ns] (time)
data['stop_date'] = pd.to_datetime(data['stop_date'], errors='coerce')
data['stop_time'] = pd.to_datetime(data['stop_time'], format='%H:%M:%S', errors='coerce').dt.time
```
